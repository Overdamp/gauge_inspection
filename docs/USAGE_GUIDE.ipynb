## Step 9: Run Batch Processing

### 9.1 Configure Batch Processing Paths

```python
import os

# Define directories
input_folder = "/home/engineer00/winworkspace/PTTEP/Objectdetection/crop_gauge_result/analog-gauge"
output_folder = "/home/engineer00/winworkspace/PTTEP/AnalogGaugeReading/gauge-win-v3/results"

print("🎯 BATCH PROCESSING CONFIGURATION\n")
print(f"Input Folder:  {input_folder}")
print(f"Output Folder: {output_folder}")

# Create output directory if it doesn't exist
os.makedirs(output_folder, exist_ok=True)
print(f"\n✅ Output folder ready: {output_folder}")
```

### 9.2 Run Batch Inference

```python
import cv2
import glob
import time
import os

input_folder = "/home/engineer00/winworkspace/PTTEP/Objectdetection/crop_gauge_result/analog-gauge"
output_folder = "/home/engineer00/winworkspace/PTTEP/AnalogGaugeReading/gauge-win-v3/results"

# Find all images
valid_extensions = ('*.jpg', '*.jpeg', '*.png', '*.bmp')
image_paths = []
for ext in valid_extensions:
    image_paths.extend(glob.glob(os.path.join(input_folder, ext)))
    image_paths.extend(glob.glob(os.path.join(input_folder, ext.upper())))

if not image_paths:
    print(f"❌ No images found in: {input_folder}")
else:
    print(f"📸 BATCH PROCESSING\n")
    print(f"Total images to process: {len(image_paths)}")
    print(f"\n🔄 Starting batch processing...\n")
    
    success_count = 0
    start_time = time.time()
    
    for idx, img_path in enumerate(image_paths, 1):
        filename = os.path.basename(img_path)
        
        # Load image
        img = cv2.imread(img_path)
        if img is None:
            logger.error(f"[{idx}/{len(image_paths)}] Cannot read: {filename}")
            continue
        
        # Process
        result = analog_task.execute(img, filename=filename)
        
        if result and "value" in result:
            value = result.get("value", 0.0)
            unit = result.get("unit", "")
            r2_score = result.get("r2_score", 0.0)
            
            logger.info(f"[{idx}/{len(image_paths)}] ✅ {filename}: {value:.2f} {unit} (R²: {r2_score:.2f})")
            success_count += 1
            
            # Save debug image
            if result.get("debug_img") is not None:
                out_path = os.path.join(output_folder, f"result_{filename}")
                cv2.imwrite(out_path, result["debug_img"])
        else:
            logger.warning(f"[{idx}/{len(image_paths)}] ❌ Failed: {filename}")
            if result and result.get("debug_img") is not None:
                out_path = os.path.join(output_folder, f"failed_{filename}")
                cv2.imwrite(out_path, result["debug_img"])
    
    elapsed = time.time() - start_time
    
    print(f"\n" + "="*50)
    print(f"✅ BATCH PROCESSING COMPLETED")
    print(f"="*50)
    print(f"Processed: {len(image_paths)} images")
    print(f"Successful: {success_count}/{len(image_paths)}")
    print(f"Success Rate: {(success_count/len(image_paths)*100):.1f}%")
    print(f"Time Elapsed: {elapsed:.2f} seconds")
    print(f"Avg Time/Image: {(elapsed/len(image_paths)):.2f} seconds")
    print(f"\nResults saved to: {output_folder}")
```
